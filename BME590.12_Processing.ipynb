```python
# Basic Packages 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Writing Excel File
import xlsxwriter
```

```python
df_label = pd.read_excel("data.xlsx")
df_label.columns
```

```python
df_label['subject'] = df_label['subject'].str.replace("-", "")
df_label['subject_num'] = df_label['subject'].str.extract('(\d+)', expand=False)
df_label = df_label[['subject_num', 'label']]
df_label.head(25)
```

```python
df_value = pd.read_excel("Full_scan3.xls")
df_value.columns = ['subject', 'left_top', 'left_mid', 'left_bot', 'right_top', 'right_mid', 'right_bot']
df_value['subject_num'] = df_value['subject'].str.extract('(\d+)', expand=False)
df_value.columns
```

```python
df_value.head(50)
```

```python
df_value_labels = pd.merge(df_value, df_label, on = 'subject_num', how = 'inner')
df_value_labels.head(60)
```

```python
df_value_labels.to_excel("labeled_full_scan3.xlsx", index = False) 
```
